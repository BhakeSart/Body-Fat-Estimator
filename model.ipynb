```python
def plotdist(col):
    plt.figure(figsize=(12,5))
    sns.distplot(df["BodyFat"],color="red",hist=False,label="BodyFat")
    sns.distplot(df[col],color="blue",hist=False,label=col)
    plt.legend()
    plt.show()
    
cols=list(df.columns)
for i in cols:
    print(f"Distribution Plots for {i} feature is shown")
    plotdist(i)
    print("-"*100)
    
    
```

```python
import scipy.stats as stats
```

```python
def drawplots(df,col):
    plt.figure(figsize=(15,7))
    plt.subplot(1,3,1)
    plt.hist(df[col],color="blue")
    
    plt.subplot(1,3,2)
    stats.probplot(df[col],dist="norm",plot=plt)
    
    plt.subplot(1,3,3)
    sns.boxplot(df[col],color="blue")
    plt.show()
    
cols=list(df.columns)
for i in range(len(cols)):
    print(f"Distribution Plots for {i} feature is shown")
    drawplots(df,cols[i])
    print("-"*100)
    
```

### Outlier Check 

```python
upperlimit=[]
lowerlimit=[]
for i in df.columns:
    upperlimit.append(df[i].mean()+(df[i].std())*4)
    lowerlimit.append(df[i].mean()+(df[i].std())*4)
```

```python
cols=list(df.columns)
j=0
for i in range(len(cols)):
    temp=df.loc[(df[cols[i]]>upperlimit[j])&(df[cols[i]]<lowerlimit[j])]
```
